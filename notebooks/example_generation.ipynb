```python
import statistics

num_trials = 100
data_fields = 2
gen_size = 5

# function that helps 
def generate_and_tag(data_len:int, gen_size:int) -> list[bytearray]: 

    generation = generate_symbols_random(data_len, gen_size)
    tagged_gen = generate_all_tags(generation)
    return tagged_gen
```

```python

def monte_carlo_notebook(trials:int):
    ic("Monte Carlo for", data_fields, gen_size)
    assert trials > 0

    accepts = []

    for trial in range(trials):
        generation = generate_and_tag(data_fields,gen_size)
        accepts.append(test_orth_generation(generation))

    ic.enable()
    prob = statistics.mean(accepts)
    std = statistics.stdev(accepts) / (trials ** 0.5) if len(accepts) > 1 else 0

    total_accepted = accepts.count(1)
    #ic(total_accepted)
    #ic(accepts)
    #ic(prob,std)

    ic.disable()
    print(f"Acceptance probability: {prob:.6f} ± {std:.6f} over {trials} trials")

    return prob, std
```

```python
num_trials = 10000
data_fields = 2
gen_size = 7

prob, std = monte_carlo_notebook(num_trials)
ic(prob, std)
ic(num_trials)
```
